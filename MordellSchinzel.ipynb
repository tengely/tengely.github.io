# Theorem 2.
## $n$ is even:

```python
P.<u,v,A,Q>=QQ[]
R.<U,V>=QQ[]
L=Set([-1,0,(4*v-3)/5,-(4*v-3)/5,1/25*(2*v+10*u+1)*(2*v+1),1/25*(-8*v^2-3*v-25*u+13),1/25*(2*v-10*u+1)*(2*v+1),1/25*(-8*v^2-3*v+25*u+13)])
K=Set([A,A+Q,A+2*Q,A+3*Q,A+4*Q])
L4=[list(k) for k in L.subsets(4)]
K4=[list(k) for k in K.subsets(4)]
def EF(k):
    if k in [0,-1]:
        return k
    else:
        return R(k(u=U,v=V)).subs(Ivar[0])

for w1 in L4:
    for v1 in K4:
        Per=Permutations(v1).list()
        for w2 in Per:
            I=([v^2-5*u^2-4]+[w1[t]-w2[t] for t in [0..3]])*P
            IE=I.elimination_ideal([A,Q])
            IG=IE.groebner_basis()
            if IG[0]!=1:
                IGR=[R(s(u=U,v=V)) for s in IG]
                Ivar=(IGR*R).variety()
                if len(Ivar)>0:
                    w1a=[EF(s) for s in w1]
                    print(Ivar,w1a,w2)
```

## $n$ is odd:

```python
P.<u,v,A,Q>=QQ[]
R.<U,V>=QQ[]
L=Set([-1,0,(4*v+3)/5,-(4*v+3)/5,1/25*(2*v+10*u-1)*(2*v-1),1/25*(-8*v^2+3*v+25*u+13),1/25*(2*v-10*u-1)*(2*v-1),1/25*(-8*v^2-3*v-25*u+13)])
K=Set([A,A+Q,A+2*Q,A+3*Q,A+4*Q])
L4=[list(k) for k in L.subsets(4)]
K4=[list(k) for k in K.subsets(4)]

def EF(k):
    if k in [0,-1]:
        return k
    else:
        return R(k(u=U,v=V)).subs(Ivar[0])

for w1 in L4:
    for v1 in K4:
        Per=Permutations(v1).list()
        for w2 in Per:
            I=([v^2-5*u^2-4]+[w1[t]-w2[t] for t in [0..3]])*P
            IE=I.elimination_ideal([A,Q])
            IG=IE.groebner_basis()
            if IG[0]!=1:
                IGR=[R(s(u=U,v=V)) for s in IG]
                Ivar=(IGR*R).variety()
                if len(Ivar)>0:
                    w1a=[EF(s) for s in w1]
                    print(Ivar,w1a,w2)
```

# Theorems 4 and 5.
### Here ncpus=128 should be modified based on the number of cores.

```python
P.<X>=QQ[]

var('x,y')

def alldiv(n):
    d0=n.divisors()
    return d0+[-k for k in d0]

@parallel(ncpus=8)
def KJ(n,a1,a2):
    N=-(4*a2^3 - 3*a2*n^2 + n^3 - 18*a1*a2 + 9*a1*n + 27)
    DN=alldiv(N)
    pol1=3*x+3*y+(2*a2 - n)
    pol2=9*x^2-9*x*y+9*y^2+(3*a2 + 3*n)*x+(3*a2 + 3*n)*y+(-2*a2^2 - a2*n + n^2 + 9*a1)
    SOL=[]
    for d1 in DN:
        d2=N//d1
        y0=solve(pol1==d1,y)[0].rhs()
        POL2=P((pol2(y=y0))(x=X))-d2
        s=POL2.roots(ring=ZZ)
        if len(s)>0:
            for k in s:
                SOL=SOL+[(k[0],y0(x=k[0]))]
    return SOL
```

## Theorem 4: here $a_1=0, a_2=-1$
